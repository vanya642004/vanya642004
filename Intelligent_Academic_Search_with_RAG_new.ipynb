```python
import os
from pypdf import PdfReader

pdf_folder = "/content/drive/MyDrive/project/"  # Change this to your folder name

all_texts = []

for file in os.listdir(pdf_folder):
    if file.endswith(".pdf"):
        pdf_path = os.path.join(pdf_folder, file)
        reader = PdfReader(pdf_path)

        text = ""
        for page in reader.pages:
            text += page.extract_text() + "\n"

        all_texts.append(text)

print("✅ Extracted text from all PDFs successfully!")
```

```python
import os
import glob
from langchain.document_loaders import PyPDFLoader

# Path to your folder containing PDFs
pdf_folder = "/content/drive/MyDrive/project/"

# Load all PDFs in the folder
documents = []
pdf_files = glob.glob(os.path.join(pdf_folder, "*.pdf"))

for pdf in pdf_files:
    loader = PyPDFLoader(pdf)
    documents.extend(loader.load())

print(f"Loaded {len(documents)} documents from PDFs")
```

```python
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings

# Load embeddings model
embedding_function = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# Store extracted text in a vector database
vector_db = Chroma.from_documents(documents, embedding_function)
print("✅ Documents stored in ChromaDB!")
```

```python
from langchain.chains import RetrievalQA
from langchain_huggingface import HuggingFaceEndpoint

# Load Falcon-7B
llm = HuggingFaceEndpoint(
    endpoint_url="https://api-inference.huggingface.co/models/tiiuae/falcon-7b-instruct",
    temperature=0.8,
    max_length=1024
)

# Create RAG pipeline
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vector_db.as_retriever(),
    chain_type="stuff"
)

# Ask a question (using your PDFs)
query = "why do we do dimentionality reduction"
response = qa_chain.run(query)
print("📖 AI Answer:", response)
```

```python
import streamlit as st
```

```python
st.title("📚 AI-Powered Education Search Engine")
```

```python
uploaded_files = st.file_uploader("Upload PDFs", type=["pdf"], accept_multiple_files=True)
```

```python
if uploaded_files:
    # Save PDFs locally
    pdf_folder = "pdfs"
    os.makedirs(pdf_folder, exist_ok=True)

    for file in uploaded_files:
        with open(os.path.join(pdf_folder, file.name), "wb") as f:
            f.write(file.getbuffer())

    # Load PDFs and extract text
    documents = []
    pdf_files = glob.glob(os.path.join(pdf_folder, "*.pdf"))

    for pdf in pdf_files:
        loader = PyPDFLoader(pdf)
        documents.extend(loader.load())

    # Convert to embeddings and store in ChromaDB
    embedding_function = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
    vector_db = Chroma.from_documents(documents, embedding_function)

    # Load Falcon-7B
    llm = HuggingFaceEndpoint(
        endpoint_url="https://api-inference.huggingface.co/models/tiiuae/falcon-7b-instruct",
        temperature=0.8,
        max_length=1024
    )

    # Create RAG pipeline
    qa_chain = RetrievalQA.from_chain_type(llm=llm, retriever=vector_db.as_retriever(), chain_type="stuff")

    # Search Box
    query = st.text_input("Ask a question based on the uploaded documents:")

    if query:
        response = qa_chain.run(query)
        st.write("📖 **Answer:**", response)
```
